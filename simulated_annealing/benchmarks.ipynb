```python
import time
import numpy as np
```

```python
# Testing how much faster it is to sum over necessary columns only.
rand_array = np.random.rand(5000,100)
rand_axis = np.random.randint(100, size=2)

cycles = 10000

time_start = time.perf_counter()
for _ in range(0, cycles):
    rand_sum = np.sum(rand_array, axis=0)
print(time.perf_counter() - time_start)
time_start = time.perf_counter()
for _ in range(0,cycles):
    rand_sum = np.sum(rand_array[:,rand_axis], axis=0)
print(time.perf_counter() - time_start)
```
